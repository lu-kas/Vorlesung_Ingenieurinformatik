```python
import cv2
import numpy as np
import random
```

```python
im = cv2.imread("logo_uni.bmp", cv2.IMREAD_GRAYSCALE)
```

```python
for i in range(0,500):
    for j in range(0,500):
        if im[i,j] == 255:
            im[i,j] = random.uniform(150,255)
        else:
            im[i,j] = random.uniform(0,149)

            
np.savetxt("im_original.csv", im, delimiter=",",fmt='%i')
```

```python
import matplotlib.pyplot as plt

plt.imshow(im, cmap='binary_r')
plt.show()
```
