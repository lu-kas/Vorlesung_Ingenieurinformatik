## For Loop

`for` loops are traditionally used when you have a block of code which you want to repeat a fixed number of times. The structure of a `for` loop consists of a iteration variable, a set of elements, which will be looped over, and the loop body. The body is the part that will be repeated for each element of the iteration set.

### Code Blocks
Many other programming languages use brackets to indicate groups of code, like the body of a for loop. Python uses indentations to group lines of code. Use a tab to create a new block.

### Comments
To add comments to your code, use the `#` character. Every text after it till the end of line will be ignored and not executed.

````python
# Example 1

# Goal: Print out the numbers 0,1,2
# iteration variable: number
# iteration set: [0,1,2] generated by the range-function
for number in range(3):
    print(number)
````

````python
# Example 2

# Print out the elements of the list movies

movies = ["Escape from LA", "Pulp Fiction", "Pumping Iron"]
for j in movies:
    print(j)
````

````python
# Example 3

# Print out elements of two lists

movies = ["Escape from LA", "Pulp Fiction", 
          "Pumping Iron", "Escape from NY"]
numbers = [11,12,13,14]

for i in range(len(movies)):
    print("iter: {}, movies: {}, numbers: {}".format(i, movies[i], 
                                                     numbers[i]))
````

````python
# Example 4

# Print out elements of two lists, now in reverse order

movies = ["Escape from LA", "Pulp Fiction", 
          "Pumping Iron", 'Escape from NY']
numbers = [21,22,23,24]

for k in range(len(movies)-1, -1, -1):
    print("iter: {}, movies: {}, numbers: {}".format(k, movies[k], 
                                                     numbers[k]))
````

A `for` loop can also be used to read in a text file line by line. Here the iteration set consists out of individual lines of a text file.
To open a file, use the `open(filename, mode)` function. Mode is either `r` (read), `w` (write), or `a` (append to a file). Once finished with the file access, the file should be closed with the `file.close()` function.

````python
# open a file for reading, the file can be accessed via the 
# data_file variable
data_file = open('data.txt', 'r')

# loop over each line, the content of the current line is 
#stored in the line variable 
for line in data_file:
    # print the content of the line
    print(line)

# close the file
data_file.close()
````

## String Manipulation
A very important task when working with strings is string manipulation, for which Python has sophisticated built-in functions.

* `str.split(separator)`: Split a string using the separator character into a list of strings
* `str.strip()`: Return a copy of the string with leading and tailing characters (whitespace, newline, tab) removed
* `str.replace(old, new)`: Return a copy of string with all occurrences of substring old replaced by new.

````python
# For example:
data_line = " 1,8; 2,5; 42,2\n"
print("example data line: ", data_line)

data_line_replace = data_line.replace(',','.')
print("replaced ',' by '.': ", data_line_replace)

data_line_strip = data_line_replace.strip()
print("strip: ", data_line_strip)

data_list_split = data_line_strip.split(';')
print("split by ';': ", data_list_split)
````

## Module glob
It is in the nature of data analysis that a large number of data is often involved. To analyse a huge amount of data files, one should loop through each file automatically.
The glob module finds all the pathnames matching a specified pattern and returns the file name.

### Modules in Python
Although Python itself has only few build-in funcitons, its functionality can be extended with modules. To load a module, the `import` statement is used, followed by the module name. The functions provided by the loaded module are accessable as `modulename.functionname`.

````python
# Example 1

import glob

# Print each file in current directory
for file in list(glob.glob('*')):
    print(file)
````

````python
# Example 2

# Print each file with the extension '.log' in each subdirectory.
# Here * is a wild card, that matches any characters
for file in sorted(list(glob.glob('weather_per_day/*.log'))):
    print(file.split('/'))
````

## If Statement

In many cases the execution of commands is not fixed, but depends on a condition. For example, if you want to sum only positive numbers out of a set of integers, you would have to check if the current iteration value is positive. Depending on the outcome of this evaluation different code blocks will be executed. This is represented with `if` statements in Python.

Python supports the following logical conditions from mathematics:
- Equals: `a == b`
- Not Equals: `a != b`
- Less than: `a < b`
- Less than or equal to: `a <= b`
- Greater than: `a > b`
- Greater than or equal to: `a >= b`

````python
# Example:

sum_positiv = 0
for i in [3,8,-9,1,-3,8,3,-8,0,-4,-1]:
    if (i < 0 and i < -5) :
        print("found a {} negative number: {}".format(i,i+1))
        print("Test1 " + str(i) + " Test")
        sum_positiv = sum_positiv + i

print("sum of numbers: {}".format(sum_positiv))
````
