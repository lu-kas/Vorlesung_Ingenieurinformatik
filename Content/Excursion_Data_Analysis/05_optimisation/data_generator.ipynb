```python
import numpy as np
import random
import matplotlib.pyplot as plt

for i in range(0,11):
    x = np.linspace(0,10, 100)
    y = np.zeros_like(x)

    term1 = random.randint(1,5)
    term2 = random.randint(1,4)

    y[x<term2] = term1*x[x<term2]
    y[x>=term2] = term1*term2

    y += 1*np.random.random(len(y))
    
    data = np.column_stack((x,y))
    
    f = open('data/simulation_{:02}'.format(i),'a')
    f.write('# x-values, y-values \n')
    np.savetxt(f,data, fmt='%.2f', delimiter=', ')

    f.close()
    
    plt.plot(data[:,0], data[:,1])
```
