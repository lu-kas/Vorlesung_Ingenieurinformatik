# Skalierung

Die Wahl der richtigen Skalierung, d.h. des Darstellungsbereichs und der Darstellungsbasis, ist beim Vergleich von Daten wichtig. 

```python
import matplotlib.pyplot as plt
```

```python
# Beispieldatensätze

x = []
y0 = []
y1 = []
for i in range(20):
    cx = i * 0.5
    x.append( cx )
    y0.append( 5*cx )
    y1.append( cx**3 * 1e-1 )
```

## Darstellungsbereich

Der Darstellungsbereich kann mit den Funktionen `plt.xlim` bzw. `plt.ylim` gewählt werden. Grundsätzlich wählt matplotlib den Bereich so aus, dass alle Daten darin abgebildet werden können. Die beiden Funktionen können diese Bereiche manuell festlegen.

Beide Funktionen nehmen eine Liste mit zwei Elementen als Argument. Das erste Element enthält den minimalen Wert der Skala und der zweite den Maximalen. Soll nur einer von beiden gesetzt werden, so kann dies durch die Argumente `xmin` und `xmax` bzw. `ymin` und `ymax` erfolgen.

```python
# Vergleichsgraphik
plt.plot(x, y0)
plt.plot(x, y1);
```

```python
# Beispielplots
plt.plot(x, y0)
plt.plot(x, y1)

# Definition des darzustellenden Bereichs
plt.xlim(xmax=3)
plt.ylim([-1,10]);
```

## Basis

Neben einer linearen Darstellung, können für Graphiken z.B. logarithmische Achsen gewählt werden. Hierzu werden die `plt.xscale` bzw. `plt.yscale` Funktionen verwendet. Um die entsprechende Achse logarithmisch darzustellen, kann ihnen das Argument `'log'` übergeben werden.

```python
plt.plot(x, y0)
plt.plot(x, y1)

plt.xscale('log')
plt.yscale('log')

plt.grid()
```
