# Beschriftung

Im matplotlib-Modul stehen viele Möglichkeiten zur Beschriftung der einzelnen Graphikelemente zur Verfügung. In diesem Kapitel wird eine kleine Auswahl präsentiert.  

```python
import matplotlib.pyplot as plt
```

## Achsenbeschriftung

Die Achsen können mit den Funktionen `plt.xlabel` bzw. `plt.ylabel` beschriftet werden.

```python
x = []
y2 = []
y3 = []
for i in range(20):
    cx = -5 + i*0.5
    x.append( cx )
    y2.append( 8 - cx**2 )
    y3.append( 0.2*cx**3 - 2*cx )
```

```python
plt.plot(x, y3)

plt.xlabel('Definitionsbereich')
plt.ylabel('Wertebereich');
```

## Titel

Über der Graphik kann eine Überschrift plaziert werden. Dies erfolgt mit der `plt.title`-Funktion.

```python
plt.plot(x, y3)

plt.xlabel('Definitionsbereich')
plt.ylabel('Wertebereich')

titel = f"Polynom 3-ten Grades zwischen {min(x):.3f} und {max(x):.3f}"
plt.title(titel);
```

## Legende

Werden mehrere Datensätze gleichzeitig dargestellt, so sollte eine Legende verwendet werden, um die einzelnen Daten zu beschreiben. Dazu muss bei den Erstellen der Graphikelemente das Argument `label` mit einer entsprechenden Zeichenkette angegeben werden. Sind alle Elemente erstellt, wird mit der Funktion `plt.legend` eine Legende generiert. 

```python
plt.plot(x, y2, label='2-er Grad')
plt.plot(x, y3, label='3-er Grad')

plt.legend()

plt.xlabel('Definitionsbereich')
plt.ylabel('Wertebereich')

plt.title("Polynome");
```
